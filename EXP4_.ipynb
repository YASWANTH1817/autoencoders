```python

!pip install tensorflow matplotlib numpy opencv-python -q

import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import zipfile
from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, Conv2D, MaxPooling2D,
                                     UpSampling2D, BatchNormalization, Activation)
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from google.colab import files


print("Generating synthetic grayscale X-ray-like images for demo...")

IMG_SIZE = 128
NUM_SAMPLES = 1000

def generate_xray_like(n, anomaly=False):
    images = []
    for _ in range(n):
        img = np.zeros((IMG_SIZE, IMG_SIZE), dtype=np.float32)

        cv2.ellipse(img, (64, 64), (50, 60), 0, 0, 360, 0.6, -1)
        cv2.ellipse(img, (40, 60), (20, 30), 0, 0, 360, 0.3, -1)
        cv2.ellipse(img, (88, 60), (20, 30), 0, 0, 360, 0.3, -1)
        # Add noise
        noise = np.random.normal(0, 0.05, (IMG_SIZE, IMG_SIZE)).astype(np.float32)
        img = np.clip(img + noise, 0, 1)
        if anomaly:
            # Add bright spot (simulated anomaly)
            cx, cy = np.random.randint(30, 90), np.random.randint(30, 90)
            cv2.circle(img, (cx, cy), np.random.randint(5, 15), 1.0, -1)
        images.append(img)
    return np.array(images)

normal_images  = generate_xray_like(NUM_SAMPLES, anomaly=False)
anomaly_images = generate_xray_like(200,          anomaly=True)

normal_images  = normal_images[..., np.newaxis]
anomaly_images = anomaly_images[..., np.newaxis]


X_train, X_test_normal = train_test_split(normal_images, test_size=0.2, random_state=42)

print(f"Train shape : {X_train.shape}")
print(f"Test normal : {X_test_normal.shape}")
print(f"Test anomaly: {anomaly_images.shape}")

# ── 3. Build Autoencoder ──────────────────────────────────
def build_autoencoder(input_shape=(IMG_SIZE, IMG_SIZE, 1)):
    inputs = Input(shape=input_shape)

    # ── Encoder ──
    x = Conv2D(32, (3,3), padding='same')(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2,2))(x)          # 64x64

    x = Conv2D(64, (3,3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = MaxPooling2D((2,2))(x)          # 32x32

    x = Conv2D(128, (3,3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    encoded = MaxPooling2D((2,2))(x)    # 16x16  ← bottleneck

    # ── Decoder (symmetrical) ──
    x = Conv2D(128, (3,3), padding='same')(encoded)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = UpSampling2D((2,2))(x)          # 32x32

    x = Conv2D(64, (3,3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = UpSampling2D((2,2))(x)          # 64x64

    x = Conv2D(32, (3,3), padding='same')(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = UpSampling2D((2,2))(x)          # 128x128

    decoded = Conv2D(1, (3,3), padding='same', activation='sigmoid')(x)

    autoencoder = Model(inputs, decoded, name='Autoencoder')
    encoder     = Model(inputs, encoded, name='Encoder')
    return autoencoder, encoder

autoencoder, encoder = build_autoencoder()
autoencoder.summary()

# ── 4. Compile ────────────────────────────────────────────
autoencoder.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=['mse']
)

# ── 5. Train on Normal Images Only ───────────────────────
callbacks = [
    EarlyStopping(patience=10, restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(factor=0.5, patience=5, verbose=1)
]

history = autoencoder.fit(
    X_train, X_train,          # input == target
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    callbacks=callbacks,
    verbose=1
)

# ── 6. Plot Training Loss ─────────────────────────────────
plt.figure(figsize=(8,4))
plt.plot(history.history['loss'],     label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Autoencoder Training Loss')
plt.xlabel('Epoch'); plt.ylabel('Loss')
plt.legend(); plt.grid(True); plt.tight_layout()
plt.savefig('training_loss.png', dpi=150)
plt.show()

# ── 7. Reconstruction & Anomaly Scoring ──────────────────
def reconstruction_error(model, images):
    reconstructed = model.predict(images, verbose=0)
    # MSE per image
    errors = np.mean((images - reconstructed) ** 2, axis=(1,2,3))
    return errors, reconstructed

normal_errors,  normal_recon  = reconstruction_error(autoencoder, X_test_normal)
anomaly_errors, anomaly_recon = reconstruction_error(autoencoder, anomaly_images)

print(f"\nReconstruction Error  |  Normal  : mean={normal_errors.mean():.4f}  std={normal_errors.std():.4f}")
print(f"Reconstruction Error  |  Anomaly : mean={anomaly_errors.mean():.4f}  std={anomaly_errors.std():.4f}")

# ── 8. Set Threshold ─────────────────────────────────────
threshold = normal_errors.mean() + 2 * normal_errors.std()
print(f"\nAnomaly Threshold (μ + 2σ): {threshold:.4f}")

normal_preds  = (normal_errors  > threshold).astype(int)
anomaly_preds = (anomaly_errors > threshold).astype(int)

print(f"False Positives (normal flagged as anomaly): {normal_preds.sum()}/{len(normal_preds)}")
print(f"True  Positives (anomaly correctly flagged): {anomaly_preds.sum()}/{len(anomaly_preds)}")

# ── 9. Error Distribution Plot ────────────────────────────
plt.figure(figsize=(8,4))
plt.hist(normal_errors,  bins=40, alpha=0.7, label='Normal',  color='steelblue')
plt.hist(anomaly_errors, bins=40, alpha=0.7, label='Anomaly', color='tomato')
plt.axvline(threshold, color='black', linestyle='--', linewidth=2, label=f'Threshold={threshold:.4f}')
plt.title('Reconstruction Error Distribution')
plt.xlabel('MSE'); plt.ylabel('Count')
plt.legend(); plt.grid(True); plt.tight_layout()
plt.savefig('error_distribution.png', dpi=150)
plt.show()

# ── 10. Visualise Input vs Reconstruction ────────────────
def visualise(original, reconstructed, errors, title, n=6):
    fig, axes = plt.subplots(3, n, figsize=(n*2.5, 7))
    for i in range(n):
        axes[0,i].imshow(original[i,:,:,0],      cmap='gray'); axes[0,i].axis('off')
        axes[1,i].imshow(reconstructed[i,:,:,0], cmap='gray'); axes[1,i].axis('off')
        diff = np.abs(original[i,:,:,0] - reconstructed[i,:,:,0])
        axes[2,i].imshow(diff, cmap='hot');       axes[2,i].axis('off')
        axes[0,i].set_title(f'Input', fontsize=8)
        axes[1,i].set_title(f'Recon', fontsize=8)
        axes[2,i].set_title(f'MSE:{errors[i]:.3f}', fontsize=8)
    fig.suptitle(title, fontsize=13, fontweight='bold')
    plt.tight_layout()
    fname = title.lower().replace(' ','_') + '.png'
    plt.savefig(fname, dpi=150)
    plt.show()

visualise(X_test_normal,  normal_recon,  normal_errors,  'Normal Images — Reconstruction')
visualise(anomaly_images, anomaly_recon, anomaly_errors, 'Anomaly Images — Reconstruction')

# ── 11. Save Model ────────────────────────────────────────
autoencoder.save('xray_autoencoder.h5')
print("\nModel saved as xray_autoencoder.h5")
```
